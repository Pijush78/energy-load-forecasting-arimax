```python
# Line Plot Over Time
plt.figure(figsize=(10,6))
plt.plot(data['utc_timestamp'], data['IT_load_new'], label='Load')
plt.plot(data['utc_timestamp'], data['IT_solar_generation'], label='Solar_Generation')
plt.title('Line Plot Over Time')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Create time features
data['month'] = data.index.month
data['hour'] = data.index.hour
data['dayofweek'] = data.index.dayofweek
```

```python
# Monthly average
monthly_avg = data.groupby('month')['IT_load_new'].mean()
plt.figure(figsize=(10,4))
sns.barplot(x=monthly_avg.index, y=monthly_avg.values)
plt.title('Average Load by Month')
plt.xlabel('Month')
plt.ylabel('Load (MW)')
plt.show()
```

```python
# Hourly average
hourly_avg = data.groupby('hour')['IT_load_new'].mean()
plt.figure(figsize=(10,4))
sns.lineplot(x=hourly_avg.index, y=hourly_avg.values)
plt.title('Average Load by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Load (MW)')
plt.xticks(range(0, 24))
plt.show()
```

```python
# Distribution of Load Values
plt.figure(figsize=(10,4))
sns.histplot(data['IT_load_new'], bins=50, kde=True)
plt.title('Distribution of IT Load')
plt.xlabel('Load (MW)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation Heatmap
corr = data.corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# # Target variable
x = data[['IT_solar_generation', 'month', 'hour', 'dayofweek']]
y = data['IT_load_new']
```

```python
# Let's use the last 30 * 24 hours = 720 data points for testing
train_size = len(data) - 720
x_train, x_test = x[:train_size], x[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
# Fit SARIMAX model
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(y_train, exog=x_train, order=(2,1,2))
model_fit = model.fit(disp=False)

print(model_fit.summary())
```

```python
# Forecast same length as test set
forecast = model_fit.forecast(steps=len(x_test), exog=x_test)
```

```python
# Plot results
plt.figure(figsize=(12,6))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, forecast, label='Forecast', color='red')
plt.title('Forecast vs Actual - IT Load')
plt.xlabel('Timestamp')
plt.ylabel('IT Load')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, forecast)
rmse = mean_squared_error(y_test, forecast)

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
# Save the fitted model
import pickle

with open('model.pkl', 'wb') as f:
    pickle.dump(model_fit, f)
```
